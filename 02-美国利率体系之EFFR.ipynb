# 美国基准利率体系

> 美国主要基准利率只有两个，一个是EFFR（The effective federal funds rate）有效联邦基金利率,另一个是SOFR（Secured Overnight Financing Rate）有担保的隔夜融资利率。其他不那么重要的利率还有3个，OBFR，TGCR，BGCR。

### EFFR（The effective federal funds rate）有效联邦基金利率

FFR（federal funds rate）联邦基金利率是指存款类金融机构（银行和信用社）将超额准备金借给其他存款类金融机构的利率，资金借款时限为1天，出借资金无抵押，每一笔借款的利率都可以称为联邦基金利率FFR。

EFFR（The effective federal funds rate）有效联邦基金利率是根据FFR计算得出的基准利率，取存款类金融机构无抵押拆借交易的中位数利率，每个交易日早上9点，纽联储会公布前一个工作日联邦基金利率中间值作为EFFR，EFFR没有期限结构，均为隔夜利率。

计算方法具体为根据美国本土银行以及外国银行美国分行提供的 FR2420 表计算得出（因为国外没有统一的交易所，因此通过手工报数进行计算），选取隔夜拆借交易，按照交易量排序选取中位数，作为当天EFFR的数值。

EFFR target range（The effective federal funds rate target range）有效联邦基金利率目标范围，指的是美联储公开市场委员会（the federal open market committee,FOMC）设定的目标，其将通过公开市场操作，将EFFR 锁定在目标范围内。


### 美联储如何实现将EFFR锁定在目标范围内



主要通过IORB以及ON RRP 两种工具进行操作。

IORB（interest on reserve balance）是指准备金利率，即银行存在美联储的账户里，美联储向银行支付的利息。

ON RRP（Overnight Reverse Repurchase Agreements Award Rate）隔夜逆回购利率，指美联储向银行出售有价证券，约定在第二天再买回给银行，美联储占用一天的资金，向银行支付的利息。

IORB 为EFFR利率上限，如果EFFR超过IORB，资金就会从美联储的账户中流出，市场中资金量变大，从而导致EFFR利率下降至IORB的水平以下。
ON RRP 为EFFR利率下限，如果EFFR低于ON RRP，美联储就会开展隔夜逆回购，促使利率下限回到ON RRP的水平。

此时就会有聪明的小伙伴产生一个疑问，什么IORB和ON RRP都是美联储支付给金融机构的资金，但是一个是上限一个是下限？

因为在IORB中金融机构的资金是无限的，而在ON RRP中美联储的资金是无限的。
