# 美联储介绍

> 预测美联储加息政策对于研究资产配置至关重要，试图预测加息，我们可以从美联储开始了解。

### 成立背景
19世纪中叶，由于缺乏政府信用背书，美国深受金融恐慌困扰，1907年严重的金融危机迫使美国国会下定决心建立美国中央银行美联储。

事实上最古老的央行是瑞典的riksbank，成立于1668年，英格兰银行随后于1694年成立。

美联储成立于1913年，根据联储法，其主要职能在于最大限度刺激就业率，保证物价稳定，让长期利率保持在适当位置。职责包括，通过通过影响经济中的货币和信贷条件来执行国家的货币政策，以追求最大的就业、稳定的价格和适度的长期利率。监督和管理银行机构，确保国家银行和金融系统的安全和健全，保护消费者的信贷权利。维护金融系统的稳定，控制金融市场可能出现的系统性风险。为存款机构、美国政府和外国官方机构提供金融服务，包括在国家支付系统的运作中发挥重要作用。

在网上我们可以看到很多说法，美联储不由美国政府控股，政策独立性较强，事实并非完全如此。确实美联储分支机构的股权在当地商业银行手中，其总行控股权也较为神秘，但美联储很多决策仍受其政府影响，不仅联邦储备委员会的7名成员均有美国主席任命，并且其主席必须时常向美国总统汇报工作，美联储每年盈利都要上缴美国财政，而财政的钱也会存放在美联储，因此与其说美联储是独立的决策机构，不如说相对独立。

### 组织框架
了解美联储框架只需要了解最重要的3个概念，branch、board of governor、FOMC。

Board of Governor，美联储的董事会成员，由7个人组成，均由美国主席任命，一般任期为4年，并且任期一般是错开的，不存在7个人在同一年同时退休的情况，这也是为了保证货币政策的延续性。

FOMC，美联储公开市场操作委员会，包括12个成员，其中包括7位董事会成员，纽约联储银行的行长，以及4位其他美联储分支行的行长。FOMC成员通过投票的方式决定最终是否加息，在最新公布的FOMC会议纪要中，可以看到出席的人数众多，但事实上有投票权的只有写在前面的12个人，其它人只是去讨论的。[minutes of FOMC](https://www.federalreserve.gov/monetarypolicy/files/fomcminutes20221214.pdf)

branch，美联储跟其他银行一样，不仅有总行也有分支行，美联储系统内银行必须拿出6%的资本和盈余认购当地联储银行的股份
